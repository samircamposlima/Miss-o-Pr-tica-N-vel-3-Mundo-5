```python
def adicionar_hora_se_necessario(data):
    if isinstance(data, str) and len(data) == 8:  # Verifica se é uma string com 8 caracteres
        return data + ' 00:00'
    else:
        return data

missaoPratica['InvoiceDate'] = missaoPratica['InvoiceDate'].apply(adicionar_hora_se_necessario)
missaoPratica['InvoiceDate'] = missaoPratica['InvoiceDate'].fillna('120201226')
missaoPratica['InvoiceDate'] = missaoPratica['InvoiceDate'].replace('20201226 00:00', '26/12/2020 00:00')



print(missaoPratica)
```
